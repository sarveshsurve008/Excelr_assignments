# Hypothesis testing exercise

Q1.A F&B manager wants to determine whether there is any
significant difference in the diameter of the cutlet between two
units. A randomly selected sample of cutlets was collected from
both units and measured? Analyze the data and draw inferences
at 5% significance level. Please state the assumptions and tests
that you carried out to check validity of the assumptions.

```python
import numpy as np
import pandas as pd 
from scipy import stats
from scipy.stats import norm 
```

```python
data=pd.read_csv('C:\\Users\\SARVESH SURVE\\OneDrive\\Desktop\\data\\Cutlets (1).csv')
```

```python
data
```

```python

unit1_cutlets = data['Unit A']
unit2_cutlets = data['Unit B']


t_statistic, p_value = stats.ttest_ind(unit1_cutlets, unit2_cutlets)


print("t-statistic:", t_statistic)
print("p-value:", p_value)


if p_value < 0.05:
    print("There is a significant difference in the diameter of the cutlets between the two units.")
else:
    print("There is no significant difference in the diameter of the cutlets between the two units.")


```

Q2. A hospital wants to determine whether there is any
difference in the average Turn Around Time (TAT) of reports
of the laboratories on their preferred list. They collected a
random sample and recorded TAT for reports of 4
laboratories. TAT is defined as sample collected to report
dispatch.
Analyze the data and determine whether there is any
difference in average TAT among the different laboratories
at 5% significance level.

```python
Labtat=pd.read_csv('C:\\Users\\SARVESH SURVE\\OneDrive\\Desktop\\data\\LabTAT.csv')
```

```python
Labtat
```

```python
Labtat['Laboratory 1'].dtypes
```

```python


lab1_tat = Labtat['Laboratory 1']
lab2_tat = Labtat['Laboratory 2']
lab3_tat = Labtat['Laboratory 3']
lab3_tat = Labtat['Laboratory 3']
lab4_tat = Labtat['Laboratory 4']


f_statistic, p_value = stats.f_oneway(lab1_tat, lab2_tat, lab3_tat, lab4_tat)


print("f-statistic:", f_statistic)
print("p-value:", p_value)


if p_value < 0.05:
    print("There is a significant difference in the average TAT among the different laboratories.")
else:
    print("There is no significant difference in the average TAT among the different laboratories.")
```

```python
from scipy.stats import chi2_contingency
```

```python

data = [[50,142,131,70],
        [435,1523,1356,750]]


chi2, p_value, dof, expected =stats.chi2_contingency(data)


print("chi-squared statistic:", chi2)
print("p-value:", p_value)
print(f"Degrees of freedom = {dof}")
print("exp", expected)


if p_value < 0.05:
    print("There is a significant association between the region and the gender of the buyers.")
else:
    print("There is no significant association between the region and the gender of the buyers.")
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between the region and the gender of the buyers.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the region and the gender of the buyers.")
```

Q4.TeleCall uses 4 centers around the globe to process 
customer order forms. They audit a certain % of the 
customer order forms. Any error in order form 
renders it defective and has to be reworked before 
processing. The manager wants to check whether 
the defective % varies by centre. Please analyze the 
data at 5% significance level and help the manager 
draw appropriate inferences

```python
Telecall = pd.read_csv('C:\\Users\\SARVESH SURVE\\OneDrive\\Desktop\\data\\Costomer+OrderForm.csv')
```

```python
Telecall = Telecall.replace({'Error Free': 0, 'Defective': 1})
```

```python
Telecall
```

```python
from scipy.stats import f_oneway


center1_defects = Telecall['Phillippines']
center2_defects = Telecall['Indonesia']
center3_defects = Telecall['Malta']
center4_defects = Telecall['India']


f_statistic, p_value = f_oneway(center1_defects, center2_defects, center3_defects, center4_defects)


print("f-statistic:", f_statistic)
print("p-value:", p_value)


if p_value < 0.05:
    print("There is a significant difference in the defective percentage among the different centers.")
else:
    print("There is no significant difference in the defective percentage among the different centers.")
```
